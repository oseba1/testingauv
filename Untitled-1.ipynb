```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def detect_lines(img, threshold1 = 50, threshold2 = 150, apertureSize = 3, minLineLength = 100, maxLineGap = 10):
    """
    
    Takes in the image and uses cv2 to find the lines and return the image and the lines that it finds

    """


    # Checking if image is found
    if img is None:
        print("[INFO] Image was not found in detect_lines method in lane_detection.py")
        return

    # Makes sure that the image is the right size. It should already be 640x480
    img = cv2.resize(img, (640, 480))


    # This slices the top half of the image
    height, width = img.shape[:2]
    img = img[height//2: :]  

    # Convert to grey scale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Findes edges and then uses edge detection to get the lines
    edges = cv2.Canny(gray, threshold1, threshold2, apertureSize=apertureSize)
    lines = cv2.HoughLinesP(
        edges, 1, np.pi / 180, 100, minLineLength=minLineLength, maxLineGap=maxLineGap
    )


    return img, lines


def draw_lines(img, lines, color = (0, 255, 0)):
    """
    
    Takes in lines and the image and returns a image that has lines drawn where the lines were calculated to be.

    """
    if lines is not None:

        for line in lines:

            x1, y1, x2, y2 = line[0]
            cv2.line(img, (x1, y1), (x2, y2), color, 2)


    return img


def get_slopes_intercepts(lines):
    """
    
    Takes in the list of lines and returns the slope and intercepts of the lines.

    """

    slopes = []
    intercepts = []


    if lines is not None:

        for line in lines:
            x1, y1, x2, y2 = line[0]

            slope = (y2 - y1) / (x2 - x1) if x2 != x1 else np.inf


            intercept = y1 - slope * x1


            slopes.append(slope)
            intercepts.append(intercept)
    
    else:
        print("Lines was none in get_slopes_intercepts in lane_detection.py")


    return slopes, intercepts


def get_interesting_slope(slopes):

    if slopes is None or len(slopes) == 0:
        print(f"Slopes is None or empty. Slopes: {slopes}")

    return max( map(abs, slopes) )



# Load test image.
img = cv2.imread('image2.png')
img, lines = detect_lines(img, 300, 350, 5, 100, 150)

# Draw detected lines.
img = draw_lines(img, lines)
plt.imshow(img)
plt.show()

# Get the slope we want to use.
slopes, intercepts = get_slopes_intercepts(lines)
print(f"{get_interesting_slope(slopes)}")
```
